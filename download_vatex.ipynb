```python
from datasets import load_dataset, Dataset
from pytube import YouTube
import os
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import numpy as np
import av
import torch
```

```python
ds = load_dataset('HuggingFaceM4/vatex', split='train[:50%]')
```

```python
def download(item):
    url = item['path']
    id = item['videoID']
    start_time = item['start']
    end_time = item['end']

    yt = YouTube(url)
    path = f'train/video_{id}.mp4'
    path_cutted = f'train/cutted_{id}.mp4'
    
    if os.path.exists(path_cutted):
        return path_cutted
    try:
        yt.check_availability()
        video = yt.streams.get_lowest_resolution()
        video.download(filename=path)
        ffmpeg_extract_subclip(path, start_time, end_time, targetname=path_cutted)
        os.remove(path)
    except Exception as e:
        print(e)
        # print(f'not available: {url}')
        return None
    return path_cutted
```

```python
ds.map(download)
```
